## Sigmoid Function

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a



x = np.arange(-10., 10., 0.01)
sig = sigmoid(x)

plt.plot(x,sig)
plt.title('Sigmoid Function')
plt.xlabel('input')
plt.ylabel('output')
plt.savefig('SigmoidFunction.png', dpi = 300)
```

## Sigmoind function Derivative

```python
def sig_d(x):
    a = []
    for item in x:
        sig = 1/(1+math.exp(-item))
        val = sig*(1.0-sig)
        a.append(val)
    return a
```

```python

x = np.arange(-10., 10., 0.01)
d = sig_d(x)

plt.plot(x,d)
plt.title('Sigmoid Function Derivative')
plt.xlabel('input')
plt.ylabel('output')
plt.savefig('SigmoidFunctionDerivative.png', dpi = 300)
```
